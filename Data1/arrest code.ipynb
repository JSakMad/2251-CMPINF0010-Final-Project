```python
import pandas as pd
import matplotlib.pyplot as plt
#reading data
df = pd.read_csv('arrest-data.csv')
neighborhood_counts = df['INCIDENTNEIGHBORHOOD'].value_counts().head(10)
neighborhood_counts = neighborhood_counts.sort_values(ascending=True)

# Mapping data
plt.bar(neighborhood_counts.index, neighborhood_counts.values, color='navy')
plt.title('Lowest 10 Pittsburgh Neighborhoods by Crime Frequency')
plt.xlabel('Location')
plt.ylabel('Arrests recorded')
plt.xticks(rotation=50, ha='right')
plt.show()
```
